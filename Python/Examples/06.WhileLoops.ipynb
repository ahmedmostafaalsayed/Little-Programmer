* Print numbers from 1 to 10 طباعة الأرقام

## Break
* while True: Infinite loops التكرار اللا نهائي
* `break` تستخدم لإيقاف التكرار
* مثال: إيقاف قراءة الطلبات إذا كتب المستخدم حرف Q
* Start with empty list of `orders`
* `join` اطبع عناصر القائمة النهائية في سطر واحد

```python
orders = []						# قائمة فارغة في البداية

while True:						# Forever تكرار للأبد
	new_order = input("Enter new order (enter Q to quit): ").lower()
	if new_order == "q":		# Quit حرف الخروج من إعادة السؤال
		break					# الخروج من الحلقة التكرارية
	orders.append(new_order)	# أضف العنصر الجديد للقائمة

print(", ".join(orders))		# اعرض عناصر القائمة وبينهم فاصلة
```

Print month names of even numbered months in English.

اطبع أسماء الشهور الفردية.

```python
months = [
	"January",
	"February",
	"March",
	"April",
	"May",
	"June",
	"July",
	"August",
	"September",
	"October",
	"November",
	"December"
]

i = 0
while i < 12:
	if i % 2 == 0:
		print(months[i])
	i += 1
```

Write a program to display the number names of the digits of a number entered by user.

For example if the number is **231** then output should be **Two Three One**

اكتب الناتج في سطر واحد

```python
from num2words import num2words

number = input("Enter a number")
length = len(number)
result = []

i = 0
while i < length:
	result.append(num2words(number[i]))
	i += 1

print(" ".join(result))
```

Ask the user about two numbers `beginning` and `end` and print sum of all numbers between them, inclusive.

اسأل المستخدم عن رقمين، ثم اطبع مجموع الأرقام من الأول إلى الآخر.

```python
beginning = int(input("Enter the beginning number: "))
end = int(input("Enter the end number: "))

sum = 0
number = beginning

while number <= end:
    sum += number
    number += 1

print(f"Sum of numers between {beginning} and {end} is: {sum}")
```

Ask the user about his name, then print his name without vowels (consonants only).

اسأل المستخدم عن اسمه ثم طبع اسمه بدون الحروف المتحركة - الحروف الساكنة فقط.

```python
name = input("Enter your name: ")
vowels = "aeiou"

i = 0
while i < len(name):
	if name[i].lower() not in vowels:
		print(name[i], end='')
	
	# Move to next character

	i += 1

print()
```
